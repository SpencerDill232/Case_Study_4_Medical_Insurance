```python
df.alcohol_freq = df.alcohol_freq.fillna('Never')
```

```python
df.info()
```

```python
df1 = df.drop(columns=["person_id"], errors="ignore")
num_cols = df1.select_dtypes(include=["int64", "float64"]).columns.tolist()
cat_cols = df1.select_dtypes(exclude=["int64", "float64"]).columns.tolist()
```

```python
import matplotlib.pyplot as plt
for col in num_cols:
    plt.figure(figsize=(6,4))
    df1[col].hist(bins=30)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

```python
for col in cat_cols:
    plt.figure(figsize=(6,4))
    df1[col].value_counts().plot(kind="bar")
    plt.title(f"Counts of {col}")
    plt.xlabel(col)
    plt.ylabel("Counts")
    plt.show()
```
